# Business Understanding

## a. Apa Itu Business Understanding

Business Understanding adalah tahap pertama dalam proyek data science yang fokus untuk memahami masalah dan tujuan bisnis sebelum masuk ke tahap teknis seperti pengolahan data atau pembuatan model.
Fase ini berfungsi untuk menerjemahkan masalah bisnis menjadi proyek yang jelas dan terarah, sehingga solusi yang dihasilkan relevan dan memberi nilai nyata bagi perusahaan/organisasi.

Tahap ini juga menjadi fondasi utama dalam CRISP-DM, karena hasil akhirnya akan menentukan arah proyek secara keseluruhan.

## b. Tujuan Business Understanding


1.   Menemukan masalah utama yang dihadapi bisnis.
2.   Menentukan tujuan bisnis, misalnya menaikkan penjualan, mengurangi pelanggan yang berhenti, atau membuat proses lebih efisien.
3.   Membuat kriteria sukses yang bisa diukur, sehingga proyek jelas kapan disebut berhasil.
4.   Menyamakan pemahaman dengan stakeholder, agar semua pihak punya ekspektasi yang sama.

## c. Aktivitas Utama


1.   Menentukan tujuan bisnis → ngobrol dengan stakeholder untuk tahu apa yang ingin dicapai.Contoh: “Kenapa penjualan turun?” atau “Target apa yang harus dicapai tahun ini?”.
2.   Menganalisis kondisi saat ini → mengecek sumber daya, batasan, dan risiko.
Contoh: “Apakah data tersedia cukup lengkap?.
3.   Menerjemahkan tujuan bisnis ke tujuan teknis → menjadikan masalah bisnis ke bentuk analisis data.Contoh: dari “kurangi churn” jadi “buat model prediksi pelanggan yang akan churn.”
4.   Membuat rencana awal proyek → menyusun langkah-langkah, timeline, serta
siapa yang bertanggung jawab.
