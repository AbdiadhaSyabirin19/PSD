# Data Exploration

Data Exploration adalah bagian dari Data Understanding yang lebih fokus ke analisis awal data.
Tujuannya:

- Melihat pola, tren, dan distribusi data.

- Mencari hubungan antar variabel (misalnya: apakah jumlah transaksi memengaruhi kemungkinan churn?).

- Mengidentifikasi anomali atau outlier yang bisa mengganggu analisis.

- Metode yang biasa digunakan:

- Statistik deskriptif (mean, median, standar deviasi).

- Visualisasi data (histogram, scatter plot, boxplot, korelasi heatmap).
