# Data Exploration

## - Pengertian

Data Exploration adalah bagian dari Data Understanding yang fokus untuk mengeksplorasi data secara visual dan statistik. Tujuannya adalah menemukan pola, tren, dan hubungan antar variabel sebelum proses modeling.

EDA membantu untuk:

- Menemukan insight awal.

- Memutuskan fitur mana yang relevan.

- Mengidentifikasi anomali atau outlier yang bisa memengaruhi model.

## - Tujuan Data Exploration

- Melihat distribusi data,Contohnya: apakah variabel target (misal churn) seimbang atau tidak.

- Mendeteksi hubungan antar variabel,Contohnya: apakah usia akun berhubungan dengan kemungkinan churn.

- Menemukan anomali atau outlier,Contohnya: pelanggan dengan transaksi 100 kali dalam sebulan (kemungkinan error input).

- Membantu memilih metode analisis/modeling,Pola data bisa menentukan apakah regresi, klasifikasi, clustering, atau metode lain lebih cocok.

## - Metode dan Teknik Data Exploration

- Statistik deskriptif: mean, median, modus, min, max, standar deviasi.

- Visualisasi data:

    - Histogram → distribusi satu variabel.

    - Boxplot → deteksi outlier.

    - Scatter plot → hubungan dua variabel.

    - Heatmap → korelasi antar variabel numerik.

    - Analisis kategori: frekuensi atau proporsi untuk variabel kategorik.
